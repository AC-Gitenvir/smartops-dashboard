```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
```

```python
# Creating the dataset manually
data = {
    "Active_Players": [100000, 150000, 200000, 250000, 300000, 350000, 400000],
    "Avg_Session_Time": [35, 40, 50, 55, 60, 65, 70],
    "Purchase_Rate": [0.12, 0.15, 0.18, 0.22, 0.25, 0.28, 0.30],
    "Marketing_Spend": [20000, 30000, 40000, 50000, 60000, 70000, 80000],
    "Genre": [1, 2, 3, 1, 2, 3, 1],  # 1: Action, 2: Puzzle, 3: RPG
    "Server_Cost": [5000, 6000, 8000, 9000, 10000, 12000, 14000],
    "Budget": [150000, 220000, 300000, 360000, 430000, 500000, 580000]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
df
```

```python
X = df[["Active_Players", "Avg_Session_Time", "Purchase_Rate", "Marketing_Spend", "Genre", "Server_Cost"]]
y = df["Budget"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)

print("R² Score:", r2_score(y_test, y_pred))
print("Mean Squared Error:", mean_squared_error(y_test, y_pred))
```

```python
new_game = [[280000, 58, 0.24, 55000, 2, 9500]]  # sample new data
predicted_budget = model.predict(new_game)
print("Predicted Budget for New Game: ₹", predicted_budget[0])
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=y_test, y=y_pred, color="skyblue", s=100)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--')  # reference line
plt.xlabel("Actual Budget (₹)")
plt.ylabel("Predicted Budget (₹)")
plt.title("Actual vs Predicted Game Budgets")
plt.grid(True)
plt.show()
```
