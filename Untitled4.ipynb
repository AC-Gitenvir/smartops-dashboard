```python
from langgraph.prebuilt import create_react_agent

from langchain.tools import Tool

from langchain_google_genai import ChatGoogleGenerativeAI

from dotenv import load_dotenv

import os

load_dotenv()
GOOGLE_API_KEY = os.getenv("AIzaSyBIonKWJsorzk9p7LpdD6lg2VtZH5klkgY")


Gemini_llm = ChatGoogleGenerativeAI(
    model="gemini-1.5-flash",
    google_api_key=GOOGLE_API_KEY,
    temperature=0.3,
    convert_system_message_to_human=True
)

def summarize_notes_fn(text: str) -> str:
    prompt = (
        "Summarize the following notes into 4-6 clear bullet points:\n\n"
        f"{text}\n\nSummary:"
    )
    response = Gemini_llm.invoke(prompt)
    return response.content.strip()

summarize_notes_tool = Tool(
    name="summarize_notes",
    func=summarize_notes_fn,
    description="Use this tool to summarize long notes or text into bullet points"
)

tools = [summarize_notes_tool]

agent = create_react_agent(Gemini_llm, tools)

user_notes = """
Computer networks are collections of interconnected devices that can communicate and share resources.
They use protocols like TCP/IP to standardize communication. Types include LAN, WAN, and MAN.
Routers, switches, and hubs are common hardware components.
The Internet is a global network connecting millions of devices.
"""


input_message = {
    "role": "user",
    "content": (
        "Please summarize the following using the summarize_notes tool:\n\n" + user_notes
    )
}

print("🔍 Summary:\n")
for step in agent.stream({"messages": [input_message]}, stream_mode="values"):
    print(step["messages"][-1].content)
```
